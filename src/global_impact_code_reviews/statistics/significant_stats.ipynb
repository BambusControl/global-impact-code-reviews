How to interpert the results of the Kruskal-Wallis:
- Kruskal-Wallis is a non-parametric test that is used to determine whether there are statistically significant differences between two or more groups.
- If the p-value is less than 0.05, then we reject the null hypothesis that the medians of the groups are equal. In other words, we conclude that there is a statistically significant difference between the groups and that difference is significant.
- Is the test statistic (also called H value) value is high, then the difference is significant. 

How to interpert the results of the Eta Squared:
- Eta-squared is a measure of size effect. It is the proportion of variance in the dependent variable that is predictable from the independent variable. The value of Eta-squared ranges from 0 to 1. 
- The closer the value of Eta-squared is to 1, the stronger the relationship between the two variables. 
- The closer the value of Eta-squared is to 0, the weaker the relationship between the two variables.
